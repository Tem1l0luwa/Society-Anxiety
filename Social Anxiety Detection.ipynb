This independence among variables makes the categorical features suitable for inclusion in a multivariate model.

```python
from scipy.stats import chi2_contingency

# Function to compute Cramr's V
def cramers_v(X, y):
    confusion_matrix = pd.crosstab(X, y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    k = min(confusion_matrix.shape)  # min(num_rows, num_cols)
    return np.sqrt(chi2 / (n * (k - 1)))

# Compute correlation matrix for categorical columns
corr_matrix = pd.DataFrame(
    np.zeros((len(cat_cols), len(cat_cols))),
    index=cat_cols,
    columns=cat_cols
)

for col1 in cat_cols:
    for col2 in cat_cols:
        corr_matrix.loc[col1, col2] = cramers_v(df[col1], df[col2])

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Cramr's V Correlation Between Categorical Features")
plt.show()

corr_matrix
```

# Model Improvement

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.feature_selection import VarianceThreshold, RFE
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

import warnings
from sklearn.exceptions import DataConversionWarning

# Suppress only DataConversionWarning
warnings.filterwarnings("ignore", category=DataConversionWarning)
```

#### The regression results reveal meaningful differences in model performance across both training and test sets. Linear Regression, while simple, achieved a respectable test R of 0.7048, indicating decent predictive power with minimal overfitting. Decision Tree Regressor, despite a perfect training MSE of 0.0000, showed signs of overfitting with a lower test R of 0.5462 and the highest test MSE. In contrast, ensemble models like Random Forest and Gradient Boosting delivered stronger generalization, with Random Forest leading in test R at 0.7750 and the lowest test MAE. Gradient Boosting followed closely, offering a balanced trade-off between bias and variance. Overall, the ensemble methods outperformed the simpler models, suggesting they are better suited for capturing complex patterns in the data.

```python
models = {
    'Linear Regression': LinearRegression(),
    'Decision Tree Regressor': DecisionTreeRegressor(random_state=42),
    'Random Forest Regressor': RandomForestRegressor(random_state=42),
    'Gradient Boosting Regressor': GradientBoostingRegressor(random_state=42)
}
```

```python
# Feature Selection (Variance Threshold)
selector = VarianceThreshold(threshold=0.01)
X_train_reduced = selector.fit_transform(X_train)
X_test_reduced = selector.transform(X_test)
selected_features = X_train.columns[selector.get_support()]

print(f"Selected {len(selected_features)} features out of {X_train.shape[1]} after VarianceThreshold.")

# using for loop
results= {}
fig, axes=  plt.subplots(1,4, figsize=(22,5))

for (name, model), ax in zip(models.items(), axes.flatten()):
    # Train the model
    model.fit(X_train_reduced, y_train)

    # Predict
    train_pred = model.predict(X_train_reduced)
    test_pred = model.predict(X_test_reduced)

    # Evaluate
    train_mse = mean_squared_error(y_train, train_pred)
    test_mse = mean_squared_error(y_test, test_pred)
    test_mae = mean_absolute_error(y_test, test_pred)
    test_r2 = r2_score(y_test, test_pred)

    # Store results
    results[name] = {
        'Train MSE': train_mse,
        'Test MSE': test_mse,
        'Test MAE': test_mae,
        'Test R': test_r2
    }

    # Plot predictions vs actual
    ax.scatter(y_test, test_pred )
    ax.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
    ax.set_title(name)
    ax.set_xlabel("Actual")
    ax.set_ylabel("Predicted")

plt.tight_layout()
plt.show()

# Print metrics
print('\nRegression Metrics')
metrics_df = pd.DataFrame(results)
print(metrics_df.round(4))
```

# Hyperparameter Tuning ---- The Ridge Regression tuning plot shows how model performance varies with different regularization strengths. The R score remains consistently high between alpha values of 0.001 and 100, indicating good generalization across this range. The peak performance occurs around an alpha that yields an R of approximately 0.6886. However, when alpha reaches 1000, the model becomes overly constrained, causing a sharp drop in predictive accuracy. This suggests that moderate regularization is optimal, while excessive penalization harms the models ability to learn meaningful patterns.

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer, r2_score

# Define alpha values to test
alphas = np.logspace(-3, 3, 20)

# Set up Ridge model and GridSearchCV
ridge = Ridge()
param_grid = {'alpha': alphas}
grid = GridSearchCV(ridge, param_grid, cv=5, scoring=make_scorer(r2_score))
grid.fit(X_train, y_train)

# Extract results
mean_scores = grid.cv_results_['mean_test_score']

# Plot
plt.figure(figsize=(8, 5))
plt.plot(alphas, mean_scores, marker='o')
plt.xscale('log')
plt.xlabel('Alpha (Regularization Strength)')
plt.ylabel('Mean CV R Score')
plt.title('Ridge Regression Hyperparameter Tuning')
plt.grid(True)
plt.show()
```
